```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython import display
%matplotlib inline
df = pd.read_csv('./survey_results_public.csv')
df.head()
```

```python
num_rows = df.shape[0]
num_cols = df.shape[1]
print('The dataset has', num_rows ,'rows and', num_cols, 'columns')
```

```python
status_vals = df.Country.value_counts();
#Provide a pandas series of the counts for each Country. 

# The below should be a bar chart of the proportion first 10 countries with the highest number of participants on the forum.

(status_vals[0:10]/df.shape[0]).plot(kind="bar");
plt.title("Country count chart");
```

```python
ed_vals = df.EdLevel.value_counts();#Provide a pandas series of the counts for each FormalEducation status

# The below should be a bar chart of the proportion of individuals in your ed_vals
# if it is set up correctly.

(ed_vals/df.shape[0]).plot(kind="bar");
plt.title("Formal Education");
```

```python
prog_vals = df.LanguageHaveWorkedWith.value_counts();#Provide a pandas series of the counts for each FormalEducation status

# The below should be a bar chart of the list of the first 10 programming Languages used by developers
(prog_vals[0:10]/df.shape[0]).plot(kind="bar");
plt.title("Programming Language Used by developers");
```
